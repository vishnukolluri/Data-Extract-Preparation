```python
#Read website data using BeautifulSoup 
wd = open('/Users/vishnukolluri/Documents/VISHNU_CIS/DSC540/Netflix - Wikipedia.html', 'r', encoding='utf8')

# Use beautifulsoup on fd
soup = BeautifulSoup(wd)

# Close the file
wd.close()

# Print the contents
print(soup.prettify())
```

```python
# fetch the required table from tag wikitable
tble = soup.find("table", attrs={"class": 'wikitable float-left'})
```

```python
# Find the table data
td = tble.tbody.find_all('tr')
```

```python
# Build a list for the headings
headers = []

# Get the headings
for th in td[0].find_all('th'):
    headers.append(th.text.replace('\n', ' ').strip())
```

```python
# Fetch Data Rows
records = [[td.get_text().strip() for td in tr.findAll('td')] for tr in td]
```

```python
# Build the data frame
wiki_data = pd.DataFrame(records, columns=headers)

# View the data frame
wiki_data
```

```python
#Step 1
#Format Headers
# Replace headings
wiki_data = wiki_data.set_axis(['Year', 'Revenue in Millions of US $', 'Net Income in Millions of US $',
                     'Price per Share in US $', 'Employees', 'Paid Memberships in Millions', 
                    'Fortune 500 Rank'], axis=1)

# View the dataframe
wiki_data
```

```python
#Step 2
#Remove the bad rows with all none
# Select columns where Revenue in Millions of US $ is not na
wiki_data = wiki_data[wiki_data['Revenue in Millions of US $'].notna()]

# View the dataframe
wiki_data
```

```python
#Step 3
#Correct the coloumn Fortune 500 rank remove #
wiki_data['Fortune 500 Rank'] = wiki_data['Fortune 500 Rank'].map(lambda x: x.lstrip('#-'))

# View the dataframe
wiki_data
```

```python
#Step 4
#convert Year to Period
wiki_data['Year'] = pd.to_datetime(wiki_data['Year']).dt.to_period('A-DEC')

wiki_data['Year']
```

```python
#Step 5
#Set period as index
wiki_data = wiki_data.set_index(['Year'])

# View the df
wiki_data
```
