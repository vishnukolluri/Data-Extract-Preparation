```python
#Changing dataset that contains time series data.
#Convert between string and date time
ts = pd.read_excel("/Users/vishnukolluri/Downloads/Week7and8data/BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx")
```

```python
ts.head()
```

```python
type(ts['Timestamp'])
```

```python
day_of_week = ts['Timestamp'].apply(lambda x: x.strftime('%A'))
# Adding day_of_week column in dataframe
ts.insert(loc=1,column='Day_of_Week',value=day_of_week)
ts.sample(5)
```

```python
print(f" Data collected is between date range {ts['Timestamp'].min()} - {ts['Timestamp'].max()}")
```

```python
#Convert timestamps to periods and back
#Change timestap into session of the day
ds=pd.cut(ts.Timestamp.dt.hour,
               [0,6,12,18,23],
               labels=['Night','Morning','Afternoon','Evening'],
               include_lowest=True)
ds.value_counts()
```
