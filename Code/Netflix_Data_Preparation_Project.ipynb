```python
# Establish connection
conn = sqlite3.connect('mydata.sqlite')
# Netflix Kaggle query
Netflix_query = pd.read_sql_query ('''Select * From Netflix_db ''', conn)
```

```python
# Load kaggle table
Netflix_df = pd.DataFrame(Netflix_query, columns = ['Title', 'Director', 'Cast', 'Country', 'Release_Year',
                                                 'Rating', 'Duration', 'Description', 'Year_Added', 'AGE', 'Title_Year'])
```

```python
# Validate the data loaded correctly
Netflix_df.head(10)
```

```python
# Netflix Wiki query
wiki_query = pd.read_sql_query ('''Select * From Wiki_db ''', conn)
```

```python
# Load wiki table
Wiki_df = pd.DataFrame(wiki_query, columns = ['Year', 'Revenue_in_Millions', 'Net_Income_in_Millions', 'Price_per_share',
                                                 'Employees', 'Paid_Memberships_in_Millions', 'Fortune_500_Rank'])
```

```python
# Validate the data loaded correctly
Wiki_df.head(10)
```

```python
# Netflix OMDB query
OMDB_query = pd.read_sql_query ('''Select * From api_data ''', conn)
```

```python
# Load OMDB table
OMDB_df = pd.DataFrame(OMDB_query, columns = ['Title', 'Year', 'Rated', 'Released', 'Runtime', 'Genre', 'Director',
                                             'Writer', 'Actors', 'Plot', 'Language', 'Country', 'Awards', 'Metascore', 
                                             'IMDBRating', 'IMDBVotes', 'IMDBID', 'BOXOFFICE'])
```

```python
# Validate the data loaded correctly
OMDB_df.head()
```

```python
# Close the connection
conn.close()
```

```python
#Merge Neflix and apidata
# Join the two dataframes
netflix_api_df = OMDB_df.join(Netflix_df.set_index('Title'), on='Title', lsuffix='_OMDB', rsuffix='_Kaggle')
```

```python
# Validate the merge
netflix_api_df.head()
```

```python
#Merge netflix_api_df with wikidata
netflix_api_annual_df = netflix_api_df.groupby('Year_Added', as_index=False).mean()
```

```python
# Drop age & Year_Added fields
netflix_api_annual_df = netflix_api_annual_df.drop(['Release_Year', 'AGE'], axis = 1)
```

```python
# View the dataframe
netflix_api_annual_df.head(11)
```

```python
#Got the data from api for only 2021
netflix_api_df['Year_Added'].unique()
```

```python
# Join the two dataframes
df = Wiki_df.join(netflix_api_annual_df.set_index('Year_Added'), on='Year', lsuffix='_OMDB', rsuffix='_Kaggle')
```

```python
# Replace blanks with np.nan
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# View the dataframe
df.head(17)
```
